# 1. Conhecendo os Dados

```python
import pandas as pd
```

**Conjunto dos Dados**

* idade — idade do paciente em anos

* sexo — sexo do paciente (0 = feminino, 1 = masculino)

* tipo_dor — tipo de dor no peito (valores categóricos representando diferentes tipos de dor)

* pressao_arterial — pressão arterial em repouso

* colesterol — nível de colesterol sérico

* glicemia_jejum_>120 — indica se a glicemia em jejum é maior que 120 mg/dL (0 = não, 1 = sim)

* resultados_ECG — resultados do eletrocardiograma em repouso

* frequencia_cardiaca_max — frequência cardíaca máxima atingida durante o teste

* depressao_ST — depressão do segmento ST causada por exercício

* inclinacao_ST — inclinação do segmento ST no pico do exercício

* numero_vasos_fluro — número de vasos principais observados por fluoroscopia

* teste_cintilografia — resultado do exame de cintilografia

* doenca_cardiaca — indica presença ou ausência de doença cardíaca (variável alvo)

# 2. Tratamento dos Dados

```python
dados.head()
```

*  **Dropando a Glicemia por ela ter sua grande maioria Zerada e correlação muito perto de Zero**

```python
dados = dados.drop('Glicemia jejum >120', axis=1)
```

```python
dados.columns
```

*  **Transformando os dados numéricos para uma escala única entre eles - modelo de machine learning aprende melhor**

```python
from sklearn.preprocessing import StandardScaler

numericos = ['Idade', 'Sexo', 'Tipo dor', 'Pressao arterial', 'Colesterol',
       'Resultados ECG', 'Frequencia cardiaca max', 'Dor exercicio',
       'Depressao ST', 'Inclinacao ST', 'Numero vasos fluro',
       'Teste cintilografia']
scaler = StandardScaler()
dados[numericos] = scaler.fit_transform(dados[numericos])
dados.head()
```

# Separação dos Dados e divisão em treino e teste

* **Separando os dados em X e Y | Separando em Treino - 80% e Teste - 20%**

```python
x = dados.drop('Doenca cardiaca', axis=1)

y = dados['Doenca cardiaca']
```

```python
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=42)
```

# 3. Modelos Machine Learning

# Regressão Linear

* **Montando o Modelo de Regressão Linear**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, RocCurveDisplay
```

* Utilizando o Hiperparâmetro **'class_weight: balanced'** -> usado para balancear a base de dados

```python
modelo_lr = LogisticRegression(class_weight='balanced')
modelo_lr.fit(x_treino, y_treino)
```

```python
y_pred_lr = modelo_lr.predict(x_teste)
resultado_lr = classification_report(y_teste, y_pred_lr)
print(resultado_lr)
```

```python
ConfusionMatrixDisplay.from_estimator(modelo_lr, x_teste, y_teste, cmap='Blues')
plt.title('Matriz de Confusão Regressão Logística')
plt.show()
```

```python
RocCurveDisplay.from_estimator(modelo_lr, x_teste, y_teste)
plt.title('Curva ROC Regressão Logística')
plt.show()
```

# Ávore de Decisão

* **Montando o Modelo de Árvore de Decisão**

```python
from sklearn.tree import DecisionTreeClassifier
```

**Hiperparâmetros:**

* Max_depth : profundidade máxima da árvore

* Class_weight : faz o balanceamento das classes

```python
modelo_dt = DecisionTreeClassifier(max_depth=5, class_weight='balanced', random_state=123)
modelo_dt.fit(x_treino, y_treino)
```

```python
y_pred_dt = modelo_dt.predict(x_teste)
resultado_dt = classification_report(y_teste, y_pred_dt)
print(resultado_dt)
```

```python
ConfusionMatrixDisplay.from_estimator(modelo_dt, x_teste, y_teste, cmap='Blues')
plt.title('Matriz de Confusão Árvore de Decisão')
plt.show()
```

```python
RocCurveDisplay.from_estimator(modelo_dt, x_teste, y_teste)
plt.title('Curva ROC Árvore de Decisão')
plt.show()
```

# Random Forest

* **Montando o Modelo de Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
```

**Hiperparâmetros:**

* N_estimators : número de itereções que irá ter

* Max_depth : profundidade máxima da árvore

* Class_weight : faz o balanceamento das classes

```python
modelo_rf = RandomForestClassifier(n_estimators=100, max_depth=5, class_weight='balanced', random_state=123)
modelo_rf.fit(x_treino, y_treino)
```

```python
y_pred_rf = modelo_rf.predict(x_teste)
resultado_rf = classification_report(y_teste, y_pred_rf)
print(resultado_rf)
```

```python
ConfusionMatrixDisplay.from_estimator(modelo_rf, x_teste, y_teste, cmap='Blues')
plt.title('Matriz de Confusão Random Forest')
plt.show()
```

```python
RocCurveDisplay.from_estimator(modelo_rf, x_teste, y_teste)
plt.title('Curva ROC Random Forest')
plt.show()
```

# 4. Busca dos melhores Hiperparâmetros

# GridSearch - Regressão Linear

* **Aplicando o GridSearch no modelo de Regressão Linear para tentar pegar os melhores Hiperparâmetros do modelo**

```python
from sklearn.model_selection import GridSearchCV
```

```python
modelo_lr_grid = LogisticRegression(class_weight='balanced')

grid = {
    'C' : [0.1, 1],
    'solver' : ['liblinear', 'saga'],
    'penalty' : ['l1', 'l2'],
    'max_iter' : [100,200,300]
}

grid_lr = GridSearchCV(modelo_lr_grid, grid, cv=5, scoring='recall_weighted')
grid_lr.fit(x_treino, y_treino)
```

```python
lr_best = grid_lr.best_estimator_
y_pred_lr_grid = lr_best.predict(x_teste)
resultado_lr_grid = classification_report(y_teste, y_pred_lr_grid)
print(resultado_lr_grid)
```

```python
ConfusionMatrixDisplay.from_estimator(lr_best, x_teste, y_teste, cmap='Blues')
plt.title('Matriz de Confusão Regressão Logística')
plt.show()
```

```python
RocCurveDisplay.from_estimator(lr_best, x_teste, y_teste)
plt.title('Curva ROC Regressão Logística')
plt.show()
```

# GridSearch - Random Forest

* **Aplicando o GridSearch no modelo de Random Forest para tentar pegar os melhores Hiperparâmetros do modelo**

```python
modelo_rf_grid = RandomForestClassifier(class_weight='balanced',bootstrap=True, random_state=42)

grid = {
    'n_estimators': [50,100,150],
    'max_depth': [5,8,10]
}

grid_rf = GridSearchCV(modelo_rf_grid, grid, cv=3, scoring='recall')
grid_rf.fit(x_treino, y_treino)
```

```python
best_rf = grid_rf.best_estimator_
y_pred_rf_grid = best_rf.predict(x_teste)
resultado_rf_grid = classification_report(y_teste, y_pred_rf_grid)
print(resultado_rf_grid)
```

```python
ConfusionMatrixDisplay.from_estimator(best_rf, x_teste, y_teste, cmap='Blues')
plt.title('Matriz de Confusão Random Forest')
plt.show()
```

```python
RocCurveDisplay.from_estimator(best_rf, x_teste, y_teste)
plt.title('Curva ROC Random Forest')
plt.show()
```
